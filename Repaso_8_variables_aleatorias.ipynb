Distribución Binomial
1. Enunciado del problema

Una máquina produce tornillos y cada tornillo tiene una probabilidad de 0,12 de resultar defectuoso.
Se toma una muestra de 20 tornillos al azar.

Se solicita:

1. Calcular la probabilidad de que exactamente 3 tornillos sean defectuosos.
2. Graficar la distribución binomial completa para todos los valores posibles entre 0 y 20 defectuosos.

2. Formulación del modelo probabilístico
Sea X la variable aleatoria que representa el número de tornillos defectuosos en una muestra de 20 unidades.
Cada tornillo puede ser:
	1. Defectuoso (éxito) con probabilidad p=0,12
  2. No defectuoso (fracaso) con probabilidad 1-p=0,88

Como se trata de n ensayos independientes con dos posibles resultados, la variable aleatoria X sigue una Distribución Binomial:

              X∼ Binomial ( n=20,  p=0,12)

Queremos calcular:
                        P(X=3)

```python

# EJERCICIO: Distribución Binomial
# ============================================

from scipy.stats import binom
import matplotlib.pyplot as plt

# --- Datos del problema ---
n = 20        # número de tornillos en la muestra
p = 0.12      # probabilidad de que un tornillo sea defectuoso
k = 3         # número de defectuosos que queremos calcular

# --- Cálculo de probabilidad ---
prob_k = binom.pmf(k, n, p)
print(f"La probabilidad de obtener exactamente {k} tornillos defectuosos es: {prob_k:.4f}")

# --- Distribución completa ---
x = range(n + 1)
probabilidades = binom.pmf(x, n, p)

# --- Gráfico ---
plt.figure(figsize=(10,5))
plt.bar(x, probabilidades)
plt.title("Distribución Binomial - Tornillos Defectuosos (n=20, p=0.12)")
plt.xlabel("Número de tornillos defectuosos")
plt.ylabel("Probabilidad")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Distribución de Poisson


En un peaje de autopista, se observa que en promedio 5 vehículos pasan por minuto.
El flujo vehicular se modela como una Distribución de Poisson.

Se pide:

1. Calcular la probabilidad de que exactamente 7 vehículos pasen en un minuto.

2. Graficar la distribución de Poisson para valores entre 0 y 15 vehículos.

```python
# ============================================
# Distribución de Poisson - Ejercicio Nuevo
# Peaje vehicular
# ============================================

from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parámetros
lam = 5     # vehículos por minuto
k = 7       # valor específico

# Probabilidad exacta
prob_k = poisson.pmf(k, lam)
print(f"La probabilidad de que pasen exactamente {k} vehículos es: {prob_k:.4f}")

# Distribución completa
x = range(0, 16)
probabilidades = poisson.pmf(x, lam)

# Gráfico
plt.figure(figsize=(10,5))
plt.bar(x, probabilidades, edgecolor='black')
plt.title("Distribución de Poisson (λ = 5) - Flujo Vehicular en Peaje")
plt.xlabel("Número de vehículos por minuto")
plt.ylabel("Probabilidad")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Interpretación del gráfico

El valor más probable es alrededor de 5 vehículos por minuto, coincidiendo con
λ.

Valores entre 4 y 6 son los más frecuentes.

El valor 7 vehículos (calculado en el punto 1) también tiene una probabilidad considerable (~10%).

Valores altos como 12 o más son poco probables.

 Distribución Normal

La duración de un proceso de ensamblaje sigue una distribución normal con:

Media:

μ=50 minutos

Desviación estándar:

σ=8 minutos

Se pide:

Calcular la probabilidad de que el proceso dure menos de 60 minutos.

Graficar la curva normal y sombrear el área correspondiente.

```python
# ============================================
# Distribución Normal - Ejercicio Simple
# ============================================

from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 50       # media
sigma = 8     # desviación estándar
x_valor = 60  # valor a evaluar

# Probabilidad
prob = norm.cdf(x_valor, mu, sigma)
print(f"La probabilidad de que el proceso dure menos de {x_valor} minutos es: {prob:.4f}")

# --- Gráfico ---
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,5))
plt.plot(x, y, label='Distribución Normal')

# Área sombreada
x_sombra = np.linspace(mu - 4*sigma, x_valor, 300)
y_sombra = norm.pdf(x_sombra, mu, sigma)
plt.fill_between(x_sombra, y_sombra, color='lightblue')

plt.title("Distribución Normal N(50, 8) con Área P(X < 60)")
plt.xlabel("Duración del proceso (minutos)")
plt.ylabel("Densidad")
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
